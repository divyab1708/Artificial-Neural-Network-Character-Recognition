```python
import numpy as np

def convertTo2Darray(arr):
  temp=[]
  temptarget=[]
  for tup in arr:
    temp.append(tup[0][:])
    temptarget.append(tup[1][:])

  return [np.array(temp), temptarget]

NP_TRAINING_SET, TRAINING_TARGET =convertTo2Darray(dataset.TRAINING_SET)
NP_TEST_SET, TEST_TARGET =convertTo2Darray(dataset.TEST_SET) 
NP_MESSAGE = np.array(dataset.MESSAGE)

```

```python
def show(arr):
  reshaped=arr.reshape(7,5)
  sb=[]
  for row in reshaped:
    for col in row:
      sb.append('#' if col==1 else ' ')
    sb.append("\n")
  
  print("".join(sb))


    
show(NP_TRAINING_SET[0])
show(NP_TRAINING_SET[1])


```

**4) In order to use the character labels in TRAINING_SET and TEST_SET, convert them into an integer class vector using ord(), then use NumPy to convert them into 26 one-hot encoded vectors.**

Converting into ordinal values

```python
def useOrd(arr):
  temp=[]
  for letter in arr:
    temp.append(ord(letter)-65)
  return temp

TRAINING_TARGET_ORD = useOrd(TRAINING_TARGET)
TEST_TARGET_ORD = useOrd(TEST_TARGET)

print("Training ord: ",TRAINING_TARGET_ORD)
print("Test ord: ",TEST_TARGET_ORD)

```



Converting to one-hot encoded vectors

```python
def convertOneHot(arr):
  oneHot=np.zeros((len(arr),26))
  for index,letter in enumerate(arr):
    oneHot[index][letter]=1
  return oneHot

TRAINING_ONEHOT=convertOneHot(TRAINING_TARGET_ORD)
TEST_ONEHOT=convertOneHot(TEST_TARGET_ORD)

print(TRAINING_ONEHOT,"\n")
print("===============","\n")
print(TEST_ONEHOT)

```
